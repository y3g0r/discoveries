```python
import inspect
```

```python
import io
buf = io.StringIO()
```

```python
stk = inspect.stack(context=10)
```

```python
for f in reversed(range(len(stk))):
    frame = stk[f]
    # print '<function_name> from <file_name>'
    print("{} from {}".format(frame.function, frame.filename), file=buf)
    print("".join(
        # add line number and current line indicator ('>')
        "{:0>2}{} {}".format(i+frame.lineno, '>' if i == frame.index else '|', line)
        for i, line in enumerate(frame.code_context)), file=buf)
    
buf.seek(0)
print(buf.read())
```
